```python
## Convert this list into other form like-
# [Action,Adventure,Fantasy,Science Fiction]
```

```python
import ast
def convert(obj):
    Genre=[]
    for i in ast.literal_eval(obj):
        Genre.append(i['name'])
    return Genre
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
## This will give error because the object we are passing cannot be string 
## So for solving this issue , we can use literal_eval function from ast library and that will convert string indices into  integers
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df.head()
```

```python
## For Cast we need to extract the top 3 actors name
df['cast'][1]
```

```python
def convertc(obj):
    Cast=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:                                         ## Select top 3 cast 
            Cast.append(i['name'])
            counter+=1
        else:
            break
    return Cast
```

```python
df['cast']=df['cast'].apply(convertc)
```

```python
df.head(2)
```

```python
def fetch_dir(obj):
    Director=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            Director.append(i['name'])
            break
    return Director
```

```python
df['crew']=df['crew'].apply(fetch_dir)
```

```python
## For every names , we need to remove spaces , like for Sam Worthingthon , and Sam Mendes , we need to remove spaces and make it one word , because model can get confused , if name is same
```

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(' ','')for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(' ','')for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(' ','')for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
df.head(3)
```

```python
def ensure_list(x):
    if isinstance(x, list):
        return x
    elif isinstance(x, str):
        return [x]
    else:
        return []

df['tags'] = (
    df['overview'].apply(ensure_list) +
    df['genres'].apply(ensure_list) +
    df['keywords'].apply(ensure_list) +
    df['cast'].apply(ensure_list) +
    df['crew'].apply(ensure_list)
)
```

```python
## There are similar words which are repeating again and again like , accept and accepts
## So we will use stemming for these - reducing the words to its root form 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
## Example
['loved','loving','lovable']
ps.stem('loving')
```

```python
new_df['tags'][0]
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
bow_vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
## So there are no words like actor and actors, as we applied stemming 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
bow_similarity=cosine_similarity(bow_vectors)
```

```python
bow_similarity[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=5000,stop_words='english')
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
tfidf_vectors=tfidf.fit_transform(df['tags']).toarray()
tfidf_similarity=cosine_similarity(tfidf_vectors)
```

```python
## For Bow
def recommend_bow(movie):
    movie=movie.lower()
    if movie not in new_df['title'].str.lower().values:
        print("Movie Not Found")
        return 
    index=new_df[new_df['title'].str.lower() == movie].index[0]    ## For fetching the index of the movie
    distances=bow_similarity[index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]   ## As enumerate will convert it into list of tuples , and we need to sort the list based on the distances , and not the index value
    for i in movie_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend_bow('Avatar')
```

```python
new_df.iloc[0].title
```

```python
new_df[new_df['title'] =='Batman Begins'].index[0]
```

```python
## For tfidf
def recommend_tfidf(movie):
    movie=movie.lower()
    if movie not in new_df['title'].str.lower().values:
        print("Movie Not Found")
        return
    movie_index=new_df[new_df['title'].str.lower() == movie].index[0]
    distances=tfidf_similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend_tfidf('Avatar')
```

## Comparison between BoW and TF-IDF:

- BoW considers frequency of words but treats all words equally.
- TF-IDF reduces the weight of common words and improves relevance.

Conclusion:
TF-IDF performed better in generating meaningful recommendations, so it is used as the final model.

```python
## “I initially implemented Bag of Words, but it gave less relevant results due to equal weighting. Then I improved the model using TF-IDF, which assigns importance based on term frequency and rarity, resulting in more accurate recommendations.”
```
